```python
import numpy as np
import pandas as pd
from tqdm import tqdm

from soc.config import CATALOG_CSV, CATALOG_EMBEDDINGS, CATALOG_EMBEDDINGS_META
from soc.data.catalog import load_catalog
from soc.models.clip_encoder import CLIPEncoder
```

```python
df = load_catalog(CATALOG_CSV)
df.head()
```

```python
encoder = CLIPEncoder()
```

```python
embeddings = []

for image_path in tqdm(df["image_path"]):
    emb = encoder.encode_image(image_path)
    embeddings.append(emb)

embeddings = np.vstack(embeddings)
embeddings.shape
```

```python
np.save(CATALOG_EMBEDDINGS, embeddings)
df.to_csv(CATALOG_EMBEDDINGS_META, index=False)

print("Saved embeddings to:", CATALOG_EMBEDDINGS)
print("Saved metadata to:", CATALOG_EMBEDDINGS_META)
```
